# Working with Tezos Alphanet

## Docker Distribution
You can follow instructions on [Tezos Docker](http://tezos.gitlab.io/mainnet/introduction/howtoget.html#docker-images) on official site.

```bash
cd ~/tezos-dev
wget https://gitlab.com/tezos/tezos/raw/master/scripts/alphanet.sh
chmod +x alphanet.sh
```

Script `alphanet.sh` is a wrapper for to access all components of tezos and also `tezos-client`. Executing the script without parameters will show all possible ways how to call individual components

```bash
./alphanet.sh
```

## Source Distribution
To build source distribution, follow steps describe in [How to Build Tezos](source_install.ipynb). `When building tezos for alphanet, it's important to use Alphanet branche from GIT.`

With source distribution it's required to run every component individually. 

### Before starting `tezos-node`, it is required to generate identity.

```bash
~/tezos-dev/tezos/tezos-node identity generate
```

### Start of the node 

```bash
~/tezos-dev/tezos/tezos-node run
```

### Configuration of `tezos-node`
Additional configuration can be done using `tezos-node config ...` command. By calling it with `--help` parameter it is possible to list all available options.

```bash
~/tezos-dev/tezos/tezos-node config --help 2>/dev/null | col -bx
```
