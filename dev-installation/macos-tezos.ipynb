# Tezos Installation Commands

## Installation of prerequisites on MacOS

### Install Brew a Tezos prerequisites + Libraries (In case the command is not installed)

```bash
#command -v "brew">/dev/null 
# || /usr/bin/ruby -e "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/master/install)"
 
brew install \
    wget \
    pkg-config \
    hidapi \
    libev \
    telnet \
    libsodium \
    ocaml \
    opam 
```

### Checkout and Build Tezos

```bash
# Checkout of the tezos source codes
mkdir -p ~/tezos-dev 
cd ~/tezos-dev
git clone https://gitlab.com/tezos/tezos.git
cd tezos
git checkout alphanet
```

```bash
#Build and installation of tezos
export OPAMYES=true

cd ~/tezos-dev/tezos
opam init --bare -n

make build-deps 
eval $(opam env)
make
make install
```
